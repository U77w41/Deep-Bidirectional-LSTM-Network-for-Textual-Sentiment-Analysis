```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
import torch
import requests
import re
import numpy as np
import pandas as pd
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
df = pd.read_csv('../input/cryptocurrency-tweets-dataset/tweets_processed.csv')
```

```python
df = df.dropna()
```

```python
#df2 = pd.read_csv('../input/cryptocurrency-tweets-dataset/tweets_processed.csv')
#df2.info()
```

```python
df.info()
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
sentiment_score(df['Tweet'].iloc[1])
```

```python
#df['sentiment'] = df['Tweet'].apply(lambda x: sentiment_score(x[:512]))
df['sentiment'] = df['Tweet'].apply(lambda x: sentiment_score(x[:512]))
```

```python
df.to_csv('output.csv')
```
